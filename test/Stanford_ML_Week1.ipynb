```python
import numpy as np
import matplotlib.pylab as plt

mu, sigma = 0, 0.1

X = np.linspace(0,10)
a = 1
b = 2
Y = a + b * X + np.random.normal(1,0.1,X.shape)*np.random.normal(10,0.1,X.shape)

plt.figure()
plt.scatter(X, Y)
plt.show()

```

```python
def J(theta0, theta1):
    return np.sum((theta0 + X*theta1 - Y)**2)

delta = 0.1
theta0 = np.arange(-10,10,delta)
theta1 = np.arange(-10,10,delta)
T1, T2 = np.meshgrid(theta0, theta1)
# Z = J(T1,T2)
dimension = theta0.size

print(dimension)
Z = np.zeros((dimension,dimension))

for i in range(dimension):
    for j in range(dimension):
        Z[i,j] = J(T1[i,j], T2[i,j])

plt.figure()
plt.contour(T1,T2,Z)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(T1, T2, Z, color='b')
plt.show()
```
